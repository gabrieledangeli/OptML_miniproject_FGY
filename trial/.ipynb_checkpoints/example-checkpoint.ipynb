```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
initial_x = 1
initial_y = 1
```

```python
def gradient_descent(initial_y, initial_x, max_iters, gamma):
    """Gradient descent algorithm."""
    # Define parameters to store x and objective func. values
    xs = [initial_x]
    ys = [initial_y]
    x = initial_x
    y = initial_y
    fs = [initial_x*initial_y]
    for n_iter in range(max_iters):
        # update x by a gradient descent step
        x = x - gamma * y
        y = y + gamma * x
        f = x*y
        # store x and objective function value
        xs.append(x)
        ys.append(y)
        fs.append(f)
    return xs,ys,fs
```

```python
def comp_gradient_descent(initial_y, initial_x, max_iters, gamma):
    """Gradient descent algorithm."""
    # Define parameters to store x and objective func. values
    xs = [initial_x]
    ys = [initial_y]
    x = initial_x
    y = initial_y
    fs = [initial_x*initial_y]
    for n_iter in range(max_iters):
        # update x by a gradient descent step
        x = x - gamma * 1/(1+(gamma*x*y)**2)*(y+gamma*(x*y)**2)
        y = y + gamma * 1/(1+(gamma*x*y)**2)*(-x+gamma*(x*y)**2)
        f = x*y
        # store x and objective function value
        xs.append(x)
        ys.append(y)
        fs.append(f)
    return xs,ys,fs
```

```python
max_iters = 100
gamma = 0.1
xs,ys,fs = gradient_descent(initial_y, initial_x, max_iters, gamma)
```

```python
plt.plot(list(range(len(xs))),xs,label='x')
plt.plot(list(range(len(ys))),ys, label='y')
plt.plot(list(range(len(fs))),fs, label='xy')
plt.grid()
plt.title('min xy')
plt.xlabel('iteration')
plt.ylabel('xy')
plt.legend()
```

```python
max_iters = 100
gamma = 0.1
xs,ys,fs = comp_gradient_descent(initial_y, initial_x, max_iters, gamma)
```
