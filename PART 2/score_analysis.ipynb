# **Inception scores (IS)** and **Fr√©chet inception distance (FID)** over different models

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

## - **CIFAR10**

```python
save_dir = 'models'
dataset = 'cifar10'
```

### **WGAN** with **Adam** optimizer

```python
gan_type = 'WGAN'
```

```python
path = os.path.join(save_dir, dataset, gan_type)
```

```python
is_score_WGAN = np.load(path+'/Inception_Scores.npy')
fid_score_WGAN = np.load(path+'/FID_Scores.npy')
```

### **WGAN** with **Adam** and **Grandient Penalty**

```python
gan_type = 'WGAN_GP'
```

```python
path = os.path.join(save_dir, dataset, gan_type)
```

```python
is_score_WGAN_GP = np.load(path+'/Inception_Scores.npy')
fid_score_WGAN_GP = np.load(path+'/FID_Scores.npy')
```

### **WGAN** with **Competitive Gradient Descent**

```python
gan_type = 'WGAN_ACGD'
```

```python
path = os.path.join(save_dir, dataset, gan_type)
```

```python
is_score_WGAN_ACGD = np.load(path+'/Inception_Scores.npy')
fid_score_WGAN_ACGD = np.load(path+'/FID_Scores.npy')
```

## **PLOT**

```python
plt.figure(figsize=(15,5))

x_is = range(len(is_score_WGAN))
plt.subplot(1,2,1)
plt.plot(x_is, is_score_WGAN, label='InceptionScore_WGAN')
plt.plot(x_is, is_score_WGAN_GP, label='InceptionScore_WGAN (Adam and GP)')
plt.plot(x_is, is_score_WGAN_ACGD, label='InceptionScore_WGAN (ACGD)')
plt.xlabel('Iteration (*10^3)')
plt.ylabel('Inception Score')
plt.legend(loc='lower right')

x_fid = range(len(fid_score_WGAN))
plt.subplot(1,2,2)
plt.plot(x_fid, fid_score_WGAN, label='FID_WGAN')
plt.plot(x_fid, fid_score_WGAN_GP, label='FID_WGAN (Adam and GP)')
plt.plot(x_fid, fid_score_WGAN_ACGD, label='FID_WGAN (ACGD)')
plt.xlabel('Iteration (*10^3)')
plt.ylabel('Fr√©chet inception distance')
plt.legend(loc='upper right')

plt.show()
```

### **LOSSES cifar10**

```python
save_dir = 'models'
dataset = 'cifar10'
gan_type = 'WGAN'
```

```python
path = os.path.join(save_dir, dataset, gan_type)
```

```python
gen_loss = np.load(path+'/Generator_loss.npy')
disc_loss = np.load(path+'/Discriminator_loss.npy')
```

```python
plt.figure(figsize=(10,5))

x_is = range(len(gen_loss))

plt.plot(x_is, gen_loss, label='Generator loss (WGAN)', color='red')
plt.plot(x_is, disc_loss, label='Discriminator loss')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Epochs', fontsize = 18)
plt.ylabel('Loss',  fontsize = 18)
plt.legend(loc='best',  fontsize = 14)
plt.title('Loss WGAN on CIFAR10', fontsize=18)
plt.grid()

plt.savefig('loss_wgan_cifar')

plt.show()
```
